```python
import pandas as pd
import numpy as np
import pyarrow as pa
import  pyarrow.parquet as pq
```

```python
dfItens = pd.read_csv("../../Microdados_enem_2019/DADOS/ITENS_PROVA_2019.csv", sep=';',encoding='iso-8859-1')
```

```python
arrayGab = {"CN":[], "MT":[], "CH":[], "LC":[]}

lstRows = []

for subject in arrayGab.keys():
    dfSubject = dfItens[dfItens["SG_AREA"]==subject]

    blueCode = dfSubject[dfSubject["TX_COR"]=="Azul"]["CO_PROVA"].unique()[0]
    dfCodes = dfSubject[dfSubject["CO_PROVA"]==blueCode][["CO_ITEM"]]

    for code in dfSubject['CO_PROVA'].unique():

        dfSubjectFiltered = dfSubject[dfSubject["CO_PROVA"]==code][["CO_ITEM", "CO_POSICAO", "TP_LINGUA"]]

        if subject == "LC":
            for i in range(2):
                dfSubjectFiltered[]
                dfSubjectFiltered["CO_POSICAO"] = dfSubjectFiltered["CO_POSICAO"] - np.min(dfSubjectFiltered["CO_POSICAO"])

        dfCodes = dfCodes.merge(dfSubjectFiltered, on="CO_ITEM", how="left", suffixes=(None, "_"+str(code)))

    dfCodes.rename(columns= {"CO_POSICAO":"CO_POSICAO_" + str(blueCode)}, inplace=True) 
    arrayResp[subject] = dfCodes

arrayResp["MT"].loc[0,"CO_POSICAO_522"] = 15
```
